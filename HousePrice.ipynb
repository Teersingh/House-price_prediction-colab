Build a mean Regression model


```python
# The simplest way to build mean regresssion model is calculate the mean of that particular target column
```

```python
mean_regrssion=house['latestPrice'].mean
```

```python
house['mean_price']=house['latestPrice'].mean
house['mean_price'].head()
```

```python
house['latestPrice'].head()
```

```python
house['price_mean']=house['latestPrice'].mean()
```

```python
house['price_mean'].head()
```

```python
# Visualize the mean regrssion model
```

```python
# for this we will use the matplotlib library of python
plt.figure(dpi=100)
k=range(0,len(house))
plt.scatter(k,house['latestPrice'].sort_values(),color='red',label='Actual Price of house')
plt.plot(k,house['price_mean'].sort_values(),color='green',label='prediction price of house')
plt.xlabel('fitted point (Ascending)')
plt.ylabel('latestPrice')
plt.title('overall Mean')
plt.legend()
```
